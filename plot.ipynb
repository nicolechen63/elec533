```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
camera_file = open('camera.pickle', 'rb')
camera_dict = pickle.load(camera_file)

for keys in camera_dict:
    print(keys)
```

```python
fig0, ax0 = plt.subplots()
ax0.plot(camera_dict["red_mask"])
# ax0.plot(i, camera_dict["red_mask"], label="red_mask")
# ax0.set_xlabel("time")
# ax0.set_ylabel("red color count")
```

```python
speed_file = open('speed.pickle', 'rb')
speed_dict = pickle.load(speed_file)

for keys in speed_dict:
    print(keys)
```

```python
fig1, ax1 = plt.subplots()
ax1.plot(speed_dict["current_speed"])
ax1.plot(speed_dict["current_output"])
ax1.set_xlabel("frame")
# ax1.set_ylabel("out")
# ax1.legend()
# ax1.show()
```

```python
direction_file = open('direction.pickle', 'rb')
direction_dict = pickle.load(direction_file)

for keys in direction_dict:
    print(keys)
```

```python
fig2, ax2 = plt.subplots()
ax2.plot(direction_dict["current_direction"])
ax2.plot(direction_dict["current_output"])
ax2.set_xlabel("frame")
# ax1.set_ylabel("out")
# ax1.legend()
# ax1.show()
```

```python
fig3, ax3 = plt.subplots()
current_output = np.append(np.append(np.append(speed_dict["current_output"][:34],np.zeros(10)),np.append(speed_dict["current_output"][34:75],np.zeros(10))),speed_dict["current_output"][75:])
ax3.plot((current_output * 4000 + 1500000)/20000000*100, label='Speed PWM')
ax3.plot((np.array(direction_dict["current_output"]) * 3000 + 1500000)/20000000*100, label='Steering PWM')
ax3.set_xlabel("frame")
ax3.legend()
```
